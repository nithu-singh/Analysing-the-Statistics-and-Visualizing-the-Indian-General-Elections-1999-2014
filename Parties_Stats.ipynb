```python
df1 = pd.read_csv('CW1999.csv')
```

```python
a=df1.pivot_table(index='PARTYABBRE',values='TOTVOTPOLL',aggfunc=len)
```

```python
a.reset_index(inplace=True)
```

```python
m=list(a.columns)
```

```python
m[0]='PARTY_ABBREVIATION'
```

```python
m[1]=' YEAR_1999'
```

```python
a.columns=m
```

```python
df2 = pd.read_csv('CW2004.csv')
```

```python
b=df2.pivot_table(index='PARTYABBRE',values='TOTVOTPOLL',aggfunc=len)
```

```python
b.reset_index(inplace=True)
```

```python
n=list(b.columns)
```

```python
n[0]='PARTY_ABBREVIATION'
n[1]='YEAR_2004'
```

```python
b.columns=n
```

```python
df3 = pd.read_csv('CW2009.csv')
```

```python
c=df3.pivot_table(index='Party Abbreviation',values='Total Votes Polled',aggfunc=len)
```

```python
c.reset_index(inplace=True)
```

```python
o=list(c.columns)
```

```python
o[0]='PARTY_ABBREVIATION'
o[1]='YEAR_2009'
```

```python
c.columns=o
```

```python
df4 = pd.read_csv('CW2014.csv')
```

```python
d=df4.pivot_table(index='Party Abbreviation',values='Total Votes Polled',aggfunc=len)
```

```python
d.reset_index(inplace=True)
```

```python
p=list(d.columns)
p[0]='PARTY_ABBREVIATION'
p[1]='YEAR_2014'
d.columns=p
```

```python
x=pd.merge(a,b,on='PARTY_ABBREVIATION',how='outer')
```

```python
y=pd.merge(x,c,on='PARTY_ABBREVIATION',how='outer')
```

```python
z=pd.merge(y,d,on='PARTY_ABBREVIATION',how='outer')
```

```python
z.set_index('PARTY_ABBREVIATION')
```

```python
z.rename(columns={' YEAR_1999':'YEAR_1999'},inplace=True)
```

```python
z.sort_values(by=['YEAR_2014','YEAR_2009','YEAR_2004','YEAR_1999'],ascending=False,inplace=True)
```

```python
z.dropna(inplace=True)
```

```python
z.set_index('PARTY_ABBREVIATION')
```

```python
z.reset_index(inplace=True)
```

```python
del z['index']
```

```python
z.drop(0,inplace=True)
```

```python
z
```

```python
z.iloc[0:21].iplot(kind='bar', x='PARTY_ABBREVIATION',y=['YEAR_1999','YEAR_2004','YEAR_2009','YEAR_2014'],title='MAJOR POLITICAL PARTIES GROWTH/DECLINE OVER LAST FOUR ELECTIONS',xTitle='POLITICAL PARTY ABBREVIATION',yTitle='NUMBER OF REPRESENTATIVES IN ELECTION',showgrid=False)
```
