```python
a=age['TOTVOTPOLL','F']
b=age['TOTVOTPOLL','M']
```

```python
def gcd(a,b):
    if b>a:
        while b > 0:
            a, b = b, a % b
        return a
    elif a>b:
        while a > 0:
            b, a = a, b % a
        return b
    else:
        return 1
```
