```python
states=list(poll.index)
```

```python
year=list(poll.columns)
```

```python
df = pd.DataFrame(poll.stack(), columns=['rate']).reset_index()
```

```python
df
```

```python
colors = Reds9
colors = colors[::-1]
mapper = LinearColorMapper(palette=colors, low=df.rate.min(), high=df.rate.max())

source = ColumnDataSource(df)

TOOLS = "hover,save,pan,box_zoom,reset,wheel_zoom"
```

```python
p = figure(title="STATEWISE VOTER TURNOUT PATTERN ({0} - {1})".format(states[0], states[-1]),
           x_range=states, y_range=list(reversed(year)),
           x_axis_location="below", plot_width=950, plot_height=650,
           tools=TOOLS, toolbar_location='above')
```

```python
p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "8pt"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = pi / 3
p.xaxis.axis_label = 'STATES'
p.yaxis.axis_label = 'GENERAL ELECTION YEAR'
```

```python
p.rect(x="STATE", y="level_1", width=1, height=1,
       source=source,
       fill_color={'field': 'rate', 'transform': mapper},
       line_color=None)
```

```python
color_bar = ColorBar(color_mapper= mapper, major_label_text_font_size="8pt",
                     ticker= BasicTicker(desired_num_ticks=len(colors)),
                     formatter= PrintfTickFormatter(format="%d%%"),
                     label_standoff=7, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')
```

```python
p.select_one(HoverTool).tooltips = [
     ('State', '@STATE in @level_1'),
     ('Voters Turnout%', '@rate%'),
]
```
