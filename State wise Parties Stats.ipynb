```python
df1 = pd.read_csv('CW1999.csv')
```

```python
q=df1[df1['POSITION']==1].copy()
```

```python
q['PARTYABBRE'].value_counts()
```

```python
p=q.pivot_table(index=['ST_NAME',"PARTYABBRE"],values='TOTVOTPOLL',aggfunc=len)
```

```python
p
```

```python
p.columns = p.columns.get_level_values(0)
```

```python
p.columns
```

```python
p.rename(columns={"TOTVOTPOLL":1999},inplace=True)
```

```python
p.unstack(level=0).iplot(kind='bar',barmode='stack', subplots=False,width=700,margin=(40,40,85,85),yTitle='NUMBER OF SEATS',xTitle='PARTY ABBREVIATION',zTitle='STATES',title="LOKSHABHA SEAT DISTRIBUTION OF VARIOUS PARTIES AFTER GENERAL ELECTON 1999",showgrid=False)
```
