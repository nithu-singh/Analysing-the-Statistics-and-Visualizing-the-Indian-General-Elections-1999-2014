```python
a=age['Candidate Age','F']
b=age['Candidate Age','M']
```

```python
def gcd(a,b):
    if b>a:
        while b > 0:
            a, b = b, a % b
        return a
    elif a>b:
        while a > 0:
            b, a = a, b % a
        return b
    else:
        return 1
```
